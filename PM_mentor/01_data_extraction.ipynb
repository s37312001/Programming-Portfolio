```python
import pandas as pd

input_file = r"C:\Users\Yun-chiao\Downloads\期中報告\complaints.csv"
output_file = r"C:\Users\Yun-chiao\Downloads\期中報告\complaints_filtered_2025_by_product.csv"

df = pd.read_csv(input_file)

# 日期轉換
df["Date received"] = pd.to_datetime(df["Date received"], errors="coerce")

# 基本篩選
df = df[
    (df["Date received"] >= "2025-12-01") &
    (df["Date received"] <= "2025-12-31") &
    (df["Product"] == "Checking or savings account") &
    (df["Sub-product"] == "Checking account") &
    (df["Consumer complaint narrative"].notna()) &
    (df["Consumer complaint narrative"].astype(str).str.strip() != "") &
    (df["Company response to consumer"] == "Closed with monetary relief")
    ]


# 較容易處理、適合做雛型的 Sub-issue篩選出來
easy_sub_issues = [
    "Deposits and withdrawals",
    "Problem accessing account",
    "Cashing a check",
    "Bounced checks or returned payments",
    "Unable to open an account"
]

df = df[df["Sub-issue"].isin(easy_sub_issues)]

# 根據日期排序
df = df.sort_values("Date received")

# 輸出
df.to_csv(output_file, index=False, encoding="utf-8-sig")

print(f"完成，已存到：{output_file}")
print(f"篩選後筆數：{len(df)}")
```
